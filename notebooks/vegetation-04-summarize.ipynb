```python
# Compute mean annual July NDVI
july_ndvi_inside_df = (
    ndvi_inside
        .groupby(ndvi_inside.date.dt.year)
        .mean(...)
        .NDVI
        .to_dataframe()
)
print("Mean NDVI inside Gila Boundary", july_ndvi_inside_df.head())

july_ndvi_outside_df = (
    ndvi_outside
        .groupby(ndvi_outside.date.dt.year)
        .mean(...)
        .NDVI
        .to_dataframe()
)
print("Mean NDVI outside Gila Boundary", july_ndvi_outside_df.head())
# Produces a NDVI Column that represents the average NDVI for each year
```

```python
# Join inside and outside GRIC df and rename columns
july_ndvi_df = (
    july_ndvi_inside_df[['NDVI']]
    .join(july_ndvi_outside_df[['NDVI']],
          lsuffix='_Inside_GRIC',
          rsuffix='_Outside_GRIC')
)
july_ndvi_df
```
